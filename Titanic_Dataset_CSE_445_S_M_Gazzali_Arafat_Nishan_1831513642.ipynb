```python
#We can also calculate the missing values in our dataframe by this python function.
#But, it's a bit complicated than we did before using isnull() method.
def missing_values_table(df):
        # Total missing values
        mis_val = df.isnull().sum()
        
        # Percentage of missing values
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        # Print some summary information
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        
        # Return the dataframe with missing information
        return mis_val_table_ren_columns
```

```python
#Ans to the ques no 5 starting:
# Importing Classifier Modules for our model
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import numpy as np
```

```python
train_df.info()
```

```python
#Applying cross validation to know the accuracy
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
#1st Model: DecisionTreeClassifier Model
dtc = DecisionTreeClassifier()
scoring = 'accuracy'
score = cross_val_score(dtc, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
# decision mean tree Score
acc_decision_tree = round(np.mean(score)*100, 2)
print(acc_decision_tree)
```

```python
#2nd Model: RandomForestClassifier Model
rf = RandomForestClassifier(n_estimators=200)
scoring = 'accuracy'
score = cross_val_score(rf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
# Random Forest Score
acc_random_forest = round(np.mean(score)*100, 2)
print(acc_random_forest)
```

```python
#3rd Model: LogisticRegression
#Using Logistic Regression Algorithm
lr = LogisticRegression(max_iter=1000)
scoring = 'accuracy'
score = cross_val_score(lr, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
# Logistic Regression Score
acc_lr = round(np.mean(score)*100, 2)
print(acc_lr)
```

```python
#4th Model: K Nearest Neighbour
knn = KNeighborsClassifier(n_neighbors=10)
scoring = 'accuracy'
score = cross_val_score(knn, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
# Logistic Regression Score
acc_knn = round(np.mean(score)*100, 2)
print(acc_knn)
```

```python
#5th Model: Support Vector Classifier
svc = SVC(probability= True)
scoring = 'accuracy'
score = cross_val_score(svc, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
# K Nearest Neighbors Score
acc_svc = round(np.mean(score)*100, 2)
print(acc_svc)
```

```python
models = pd.DataFrame({
    'Model': ['Random Forest','Logistic Regression', 
            'Decision Tree','KNN','Support Vector Machines'],
    'Score': [ acc_random_forest,acc_lr, acc_decision_tree, acc_knn, acc_svc ]})
models.sort_values(by='Score', ascending=False)
```
